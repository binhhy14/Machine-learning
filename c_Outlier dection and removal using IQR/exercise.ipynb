```python
import pandas as pd
df = pd.read_csv("height_weight.csv")
df.head()
```

```python
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (8,4)
```

```python
plt.hist(df.height, bins = 20, rwidth = 0.8)
plt.xlabel("Height (inches)")
plt.ylabel("Count")
```

```python
plt.hist(df.weight, bins = 20, rwidth = 0.8)
plt.xlabel("weight")
plt.ylabel("Count")
```

```python
Q1_weight = df.weight.quantile(0.25)
Q3_weight = df.weight.quantile(0.75)
IQR_weight = Q3_weight -Q1_weight
```

```python
lower_weight = Q1_weight -1.5*IQR_weight
upper_weight = Q3_weight + 1.5*IQR_weight
```

```python
df3 = df[(df.weight<lower_weight)| (df.weight>upper_weight)]
df3
```

```python
df4 = df[(df.weight > lower_weight)&(df.weight< upper_weight)]
df4
```
